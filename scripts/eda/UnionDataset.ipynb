## Objetivo: 
    - El objetivo de este notebook es unir los dos dataset por medio de la clave ID para así tener solo un dataset conjunto y sea más fácil operar con los datos.

```python
df_app=pd.read_csv('../data_acquisition/application_record.csv')
df_record=pd.read_csv('../data_acquisition/credit_record.csv')
```

```python
df_record.drop_duplicates(subset="ID",inplace=True)
```

```python
df_total = pd.merge(df_app, df_record, on='ID', how='inner')
```
