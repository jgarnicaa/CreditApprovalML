```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

```

## Objetivo: 
    - El objetivo de este notebook es transformar todos los datos categoricos en datos numéricos para facilitar su operación

```python
df = pd.read_csv('../data_acquisition/general.csv')
df=df.drop(columns=['Unnamed: 0'])
le = LabelEncoder()
```

```python
### Valores únicos por columna del dataset aplicación

for i in df.columns.values:
    if i != 'AMT_INCOME_TOTAL':
        print(str(i), df[i].unique())
```

```python
binary_columns = ['CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']
for col in binary_columns:
    df[col] = le.fit_transform(df[col])
```

```python
multi_category_columns = [
    'NAME_INCOME_TYPE',
    'NAME_EDUCATION_TYPE',
    'NAME_FAMILY_STATUS',
    'NAME_HOUSING_TYPE',
    'STATUS'
]
for col in multi_category_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].fillna('Unknown')
df['OCCUPATION_TYPE'] = le.fit_transform(df['OCCUPATION_TYPE'])
```

```python
df.head()
```

```python
for i in df.columns.values:
    if i != 'AMT_INCOME_TOTAL':
        print(str(i), df[i].unique())
```

Como se ve el dataset ahora está totalmente numerizado y se puede ver el valor que toma cada variable como un valor numérico único para cada categoria.
